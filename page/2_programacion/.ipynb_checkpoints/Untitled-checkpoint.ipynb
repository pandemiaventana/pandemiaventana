```python
import pandas as pd
import numpy as np

### Para graficar

import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import matplotlib as mpl
from matplotlib.lines import Line2D
### Customizamos para tener fondos transparentes con alta resolución
mpl.rcParams['text.color'] = 'white'
mpl.rcParams['figure.dpi']= 300
mpl.rc("savefig", dpi=300)
plt.rcParams.update({
    "figure.facecolor":  (0.0, 0.0, 0.0, 0.0),
    "axes.facecolor":    (0.0, 0.0, 0.0, 0.0),
    "savefig.facecolor": (0.0, 0.0, 0.0, 0.0),
})
plt.rc('axes.spines', **{'bottom':True, 'left':True, 'right':False, 'top':False})

import seaborn as sns
```

```python
### Leemos el archivo Excel con nuestro amigo infalible, Pandas

aportes_primarias = pd.read_excel(
'https://repodocgastoelectoral.blob.core.windows.net/public/Reporte_Aportes_Abonados_PRIMARIAS_PRESIDENCIALES_PARLAMENTARIAS.xlsx',)

### Asignamos nombres a columnas

aportes_primarias.columns = aportes_primarias.iloc[6]

### Descartamos primeras dos seis filas sin información útil

aportes_primarias = aportes_primarias.drop(np.arange(7))

### Para monto acumulado total

### Agrupamos por monto acumulado por candidato ###

aportes_primarias_acum = aportes_primarias.groupby(['NOMBRE CANDIDATO/PARTIDO POLITICO', 'PACTO'], as_index=False)['MONTO'].sum()

### Agrupamos por monto acumulado por pacto ###

aportes_primarias_acum_pacto = aportes_primarias.groupby(['PACTO'], as_index=False)['MONTO'].sum()
aportes_primarias_acum_pacto['COLOR'] = ['tab:blue', 'tab:red']

### Nos quedamos solo con el apellido ###

nombres = []
for nombre in aportes_primarias_acum['NOMBRE CANDIDATO/PARTIDO POLITICO'].str.split():
    nombres += [nombre[1]]

aportes_primarias_acum['NOMBRE CANDIDATO/PARTIDO POLITICO'] = nombres
aportes_primarias_acum['COLOR'] = ['tab:blue', 'tab:blue', 'tab:red', 'tab:red', 'tab:red', 'tab:red']

### Para serie de tiempo acumulando montos por candidato
aportes_primarias['MONTO ACUMULADO'] = aportes_primarias.groupby('NOMBRE CANDIDATO/PARTIDO POLITICO')['MONTO'].transform(pd.Series.cumsum).astype(int)

### Para serie de tiempo acumulando montos por candidato
aportes_primarias['MONTO ACUMULADO POR PACTO'] = aportes_primarias.groupby('PACTO')['MONTO'].transform(pd.Series.cumsum).astype(int)
```

```python
### Código extraído desde https://matplotlib.org/stable/gallery/misc/packed_bubbles.html ###

class BubbleChart:
    def __init__(self, area, bubble_spacing=0):
        """
        Setup for bubble collapse.

        Parameters
        ----------
        area : array-like
            Area of the bubbles.
        bubble_spacing : float, default: 0
            Minimal spacing between bubbles after collapsing.

        Notes
        -----
        If "area" is sorted, the results might look weird.
        """
        area = np.asarray(area)
        r = np.sqrt(area / np.pi)

        self.bubble_spacing = bubble_spacing
        self.bubbles = np.ones((len(area), 4))
        self.bubbles[:, 2] = r
        self.bubbles[:, 3] = area
        self.maxstep = 2 * self.bubbles[:, 2].max() + self.bubble_spacing
        self.step_dist = self.maxstep / 2

        # calculate initial grid layout for bubbles
        length = np.ceil(np.sqrt(len(self.bubbles)))
        grid = np.arange(length) * self.maxstep
        gx, gy = np.meshgrid(grid, grid)
        self.bubbles[:, 0] = gx.flatten()[:len(self.bubbles)]
        self.bubbles[:, 1] = gy.flatten()[:len(self.bubbles)]

        self.com = self.center_of_mass()

    def center_of_mass(self):
        return np.average(
            self.bubbles[:, :2], axis=0, weights=self.bubbles[:, 3]
        )

    def center_distance(self, bubble, bubbles):
        return np.hypot(bubble[0] - bubbles[:, 0],
                        bubble[1] - bubbles[:, 1])

    def outline_distance(self, bubble, bubbles):
        center_distance = self.center_distance(bubble, bubbles)
        return center_distance - bubble[2] - \
            bubbles[:, 2] - self.bubble_spacing

    def check_collisions(self, bubble, bubbles):
        distance = self.outline_distance(bubble, bubbles)
        return len(distance[distance < 0])

    def collides_with(self, bubble, bubbles):
        distance = self.outline_distance(bubble, bubbles)
        idx_min = np.argmin(distance)
        return idx_min if type(idx_min) == np.ndarray else [idx_min]

    def collapse(self, n_iterations=50):
        """
        Move bubbles to the center of mass.

        Parameters
        ----------
        n_iterations : int, default: 50
            Number of moves to perform.
        """
        for _i in range(n_iterations):
            moves = 0
            for i in range(len(self.bubbles)):
                rest_bub = np.delete(self.bubbles, i, 0)
                # try to move directly towards the center of mass
                # direction vector from bubble to the center of mass
                dir_vec = self.com - self.bubbles[i, :2]

                # shorten direction vector to have length of 1
                dir_vec = dir_vec / np.sqrt(dir_vec.dot(dir_vec))

                # calculate new bubble position
                new_point = self.bubbles[i, :2] + dir_vec * self.step_dist
                new_bubble = np.append(new_point, self.bubbles[i, 2:4])

                # check whether new bubble collides with other bubbles
                if not self.check_collisions(new_bubble, rest_bub):
                    self.bubbles[i, :] = new_bubble
                    self.com = self.center_of_mass()
                    moves += 1
                else:
                    # try to move around a bubble that you collide with
                    # find colliding bubble
                    for colliding in self.collides_with(new_bubble, rest_bub):
                        # calculate direction vector
                        dir_vec = rest_bub[colliding, :2] - self.bubbles[i, :2]
                        dir_vec = dir_vec / np.sqrt(dir_vec.dot(dir_vec))
                        # calculate orthogonal vector
                        orth = np.array([dir_vec[1], -dir_vec[0]])
                        # test which direction to go
                        new_point1 = (self.bubbles[i, :2] + orth *
                                      self.step_dist)
                        new_point2 = (self.bubbles[i, :2] - orth *
                                      self.step_dist)
                        dist1 = self.center_distance(
                            self.com, np.array([new_point1]))
                        dist2 = self.center_distance(
                            self.com, np.array([new_point2]))
                        new_point = new_point1 if dist1 < dist2 else new_point2
                        new_bubble = np.append(new_point, self.bubbles[i, 2:4])
                        if not self.check_collisions(new_bubble, rest_bub):
                            self.bubbles[i, :] = new_bubble
                            self.com = self.center_of_mass()

            if moves / len(self.bubbles) < 0.1:
                self.step_dist = self.step_dist / 2

    def plot(self, ax, labels, colors):
        """
        Draw the bubble plot.

        Parameters
        ----------
        ax : matplotlib.axes.Axes
        labels : list
            Labels of the bubbles.
        colors : list
            Colors of the bubbles.
        """
        for i in range(len(self.bubbles)):
            circ = plt.Circle(
                self.bubbles[i, :2], self.bubbles[i, 2], color=colors[i])
            ax.add_patch(circ)
            ax.text(*self.bubbles[i, :2], labels[i],
                    horizontalalignment='center', verticalalignment='center')
```

```python
### Graficamos
bubble_chart = BubbleChart(area=aportes_primarias_acum['MONTO'],
                           bubble_spacing=2000)
bubble_chart.collapse()
fig, ax = plt.subplots(subplot_kw=dict(aspect="equal"))
bubble_chart.plot(
    ax, aportes_primarias_acum['NOMBRE CANDIDATO/PARTIDO POLITICO'] 
    #+ ' \n' + aportes_primarias_acum['MONTO'].astype(str)
    , aportes_primarias_acum['COLOR']
)
ax.axis("off")
ax.relim()
ax.autoscale_view()

### Información adicional
ax.set_title('Aporte total por campaña, Primarias Presidenciales 2021')
legend_elements = [Line2D([0], [0], marker='.', color='tab:blue', label='Apruebo Dignidad',
                          markerfacecolor='tab:blue', markersize=10, linestyle='None'),
                   Line2D([0], [0], marker='.', color='tab:red', label='Chile Vamos',
                          markerfacecolor='tab:red', markersize=10, linestyle='None')]
ax.legend(handles=legend_elements, loc='upper left', facecolor='white', framealpha=0, prop={'size': 6})
plt.show()
```

```python
nombres = []
for nombre in aportes_primarias['NOMBRE CANDIDATO/PARTIDO POLITICO'].str.split():
    nombres += [nombre[1]]

aportes_primarias['CANDIDATO'] = nombres

fig, ax = plt.subplots()
sns.lineplot(data=aportes_primarias, x="FECHA DE TRANSFERENCIA", y="MONTO ACUMULADO",
             hue="CANDIDATO", estimator=None)
ax.legend(facecolor='white', framealpha=0, prop={'size': 6})
ax.spines['bottom'].set_color('white')
ax.spines['top'].set_color('white') 
ax.spines['right'].set_color('white')
ax.spines['left'].set_color('white')
ax.xaxis.label.set_color('white')
ax.yaxis.label.set_color('white')
ax.tick_params(colors='white', which='both')
plt.xlabel("Fecha")
plt.ylabel("Aporte acumulado (CLP)")
ax.yaxis.set_major_formatter(mpl.ticker.StrMethodFormatter('{x:,.0f}'))
ax.set_title('Serie de tiempo de aportes por candidato, Primarias Presidenciales 2021')
plt.show()
```
